### E-Commerce Specification Document

#### Synopsis
---
The business is looking to design a pipeline that can process a dump of the previous month's records into a data warehouse.

The data becomes available by the 1st day of the month (Not necessarily always guarenteed)


#### Requirements
---
* Data must be accessible as soon as the data is available (Within 1 Day)
* A platform that serves data within the past 3 months (<= 90 days)
    * Ability to query historical data as far back as 6 months (<= 180 Days)
        * Data older than (<= 180 Days) can be archived as it gets rarely accessed
    * Ability to query raw data; ideally in parquet format, as far back as 3 years (<= 1080 Days) 
    

#### Data Retention Policies
---
| Access Type | Format  | Duration | Access Layer | Archive Layer |
|-------------|---------|----------|--------------|---------------|
| Quarterly   | Parquet | (3 months [<= 90 days])  | 	Redshift (Silver & Gold layer) |  |
| Semi-Annual | Parquet | (6 months [<= 180 days]) |	s3 (Hot Access/ Silver)		   | s3 (Deep Archive) |
| Annual      | CSV     | (1 year [<= 365 days])   |  s3 (Hot Access/ Landing)	   | s3 (Deep Archive) |
| Historic    | Parquet | (3 years [<= 1095 days]) |	s3 (Hot Access/ Bronze)		   | s3 (Deep Archive) |

#### Data Summary
---

Previous month's data will be uploaded to the following s3 bucket
 * `s3::/lp-ecommerce-bronze/event-logs/raw-ecom-csv/{year}/{month}`

[Data Format, Volume, Velocity]

*  A `csv` file `(5-9 GB)` containing data from the `previous month` will be uploaded into a s3 bucket at the beginning of the `next month`

#### Job Metadata Management

---
As a means of monitoring the pipeline, metadata about each job will need to be captured and stored to serve as current and historic indicators of the health of the pipeline.

In pursuit of a flexible and denormalized approach, AWS DynamoDB was selected to serve as the NoSQL platform for metadata management. Going with a schemaless model, building out a single table to store both job and job run metadata will ensure ease of accessibility while accommodating schema evolution.
